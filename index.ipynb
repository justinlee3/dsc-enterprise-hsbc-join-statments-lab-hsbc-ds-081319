## Connecting to the Database
Import the necessary packages and connect to the database **data.sqlite**.

```python
#Your code here
import sqlite3
conn = sqlite3.connect("data.sqlite" , detect_types=sqlite3.PARSE_COLNAMES)
c = conn.cursor()
```

## Display the names of all the employees in Boston.
Hint: join the employees and offices tables.

```python
#Your code here
c.execute("""select firstName, lastName from employees e
             join offices o
             on o.officeCode == e.officeCode
             where city == 'Boston'""").fetchall()
```

## Do any offices have no employees?
Hint: Combine the employees and offices tables and use a group by.

```python
#Your code here
c.execute("""select count (*), city from employees e
             join offices o
             on o.officeCode == e.officeCode
             group by city""").fetchall()
```

## Level Up: Display the names of each product each employee has sold

```python
# Your code here
c.execute ("""select firstName, lastName, productName from products p
              join orderdetails using (productCode)
              join orders o using (orderNumber)
              join customers c using (customerNumber)
              join employees e on e.employeeNumber == c.salesRepEmployeeNumber
              """).fetchall()
```

## Level Up: Display the Number of Products each employee has sold

```python
# Your code here
c.execute ("""select firstName, lastName, count(*) from products p
              join orderdetails using (productCode)
              join orders o using (orderNumber)
              join customers c using (customerNumber)
              join employees e on e.employeeNumber == c.salesRepEmployeeNumber
              """).fetchall()
```
